```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red', label='plt.scatter()', s=50)
plt.title('visualisasi data titik-titik', color='white', size=20)
plt.grid(True, color="blue", alpha=0.3, linestyle="--")

plt.legend()
plt.show()
```
